```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 读取数据文件

```python
file_path = "D:\毕业论文\zzy\gut_net_project\data\standardized\Crohn_Disease_combined.csv"
df = pd.read_csv(file_path)
```

# 1. 基本信息检查

```python
print("基本信息:")
print(df.info())
print("\n描述性统计:")
print(df.describe())
```

# 2. 缺失值检查

```python
missing_values = df.isnull().sum()
print("\n缺失值情况:")
print(missing_values[missing_values > 0])
```

# 3. 异常值检查：箱线图

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.iloc[:, 5:15])  # 第5到14列为菌群丰度数据
plt.title("菌群丰度的箱线图")
plt.xticks(rotation=45)
plt.show()
```

# 4. 菌群丰度分布分析：直方图

```python
df.iloc[:, 5:15].hist(bins=20, figsize=(12, 10))
plt.suptitle("菌群丰度分布的直方图")
plt.show()
```

# 5. 人口统计特征分析：年龄和BMI分布

# 年龄分布

```python
plt.figure(figsize=(10, 4))
sns.histplot(df['host_age'].dropna(), kde=True)
plt.title("年龄分布")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

# BMI分布

```python
if 'BMI' in df.columns:
    plt.figure(figsize=(10, 4))
    sns.histplot(df['BMI'].dropna(), kde=True)
    plt.title("BMI分布")
    plt.xlabel("BMI")
    plt.ylabel("Frequency")
    plt.show()
```

# 6. 性别与菌群丰度关系：按性别分组的菌群丰度箱线图

```python
if 'sex' in df.columns and df['sex'].notna().any():
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='sex', y='Bacteroides', data=df)  # 以“Bacteroides”为例，可换成其他菌群
    plt.title("性别与Bacteroides丰度关系")
    plt.show()
```

# 7. 菌群丰度的相关性热力图

```python

numeric_cols = df.select_dtypes(include=['float64', 'int64']).iloc[:, 5:15]
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm")
plt.title("菌群丰度相关性")
plt.show()
```
